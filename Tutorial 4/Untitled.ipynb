```python
def kmeans(dataSet, k):
    # initialize centroids randomly 
    numFeatures = dataSet.getNumFeatures()
    centroids = getRandomCentroids(numFeatures, k)
    
    iterations = 0
    oldCentroids = None

    # run the main k-means algorithm 
    while not shouldStop(oldCentroids, centroids, iterations):
        # save old centroids 
        oldCentroids = centroids
        iterations += 1
        
        # assign labels to each data point based on centroids 
        labels = getLabels(dataSet, centroids)
        
        # compute new centroids based on datapoint labels 
        centroids = getCentroids(dataSet, labels, k)
        
    # return final centroids
    return centroids

    
```

```python
#Returns True or False if kmens terminates either because it has run 
#the max of interations or centroids or the centroids stop changing 

def souldStop(oldCentroids, centroids, interations):
    if interations > MAX_INTERATIONS: return True
    return oldCentroids
```

```python
#Returns a label for each piece of data in the dataset 

import math 

def getLabels(dataSet, centroids):
    #for each element in the dataset, chose the closest centroid
    #make that centroid the element's label
    for element in dataSet:
        distances = [] 
        for c in centroids: 
            #euclidean distance 
            dist = math.sqrt(sum((element[i] - c[i])**2 for i in range(len(element))))
            distances.append(dist) 
            
        #index the closest centroid 
        closest = distances.index(min(distances))
        labels.append(closest)
        
    return labels     
```

```python
import random

def getCentroids(dataSet, labels, k):
    numFeatures = len(dataSet[0])

    # prepare empty centroid accumulators
    centroid_sums = [[0.0 for _ in range(numFeatures)] for _ in range(k)]
    counts = [0] * k

    # accumulate sums and counts
    for point, label in zip(dataSet, labels):
        counts[label] += 1
        for i in range(numFeatures):
            centroid_sums[label][i] += point[i]

    # compute final centroids
    centroids = []
    for i in range(k):
        if counts[i] == 0:
            # centroid has no points â€” reinitialize using random data point
            centroids.append(list(random.choice(dataSet)))
        else:
            # compute mean of assigned points
            mean = [centroid_sums[i][f] / counts[i] for f in range(numFeatures)]
            centroids.append(mean)

    return centroids

    
```
